# 使用手动制作的数据微调 ChatGLM3

````python
import torch

print(torch.__config__.show(), torch.cuda.get_device_properties(0)) # 查看torch和cuda信息
````

````python
# 定义全局变量和参数
model_name_or_path = 'THUDM/chatglm3-6b'  # 模型ID或本地路径
train_data_path = 'data/zhouyi_dataset_handmade.csv'    # 训练数据路径
eval_data_path = None                     # 验证数据路径，如果没有则设置为None
seed = 8                                 # 随机种子
max_input_length = 512                    # 输入的最大长度
max_output_length = 1536                  # 输出的最大长度
lora_rank = 8                             # LoRA秩 值原来是16
lora_alpha = 32                           # LoRA alpha值
lora_dropout = 0.05                       # LoRA Dropout率
prompt_text = ''                          # 所有数据前的指令文本
````

## 数据处理

````python
from datasets import load_dataset

dataset = load_dataset("csv", data_files=train_data_path)
print(dataset)
````

````python
from datasets import ClassLabel, Sequence
import random
import pandas as pd
from IPython.display import display, HTML

def show_random_elements(dataset, num_examples=10): # 显示数据集中的随机元素
    assert num_examples <= len(dataset), "Can't pick more elements than there are in the dataset."
    picks = []
    for _ in range(num_examples): # 随机选择num_examples个样本
        pick = random.randint(0, len(dataset)-1) # 随机选择一个样本索引
        while pick in picks:
            pick = random.randint(0, len(dataset)-1) # 确保不重复选择
        picks.append(pick) # 添加到选择列表
    
    df = pd.DataFrame(dataset[picks]) # 创建DataFrame以便显示
    for column, typ in dataset.features.items(): # 处理分类标签
        if isinstance(typ, ClassLabel): # 如果是分类标签，则转换为名称
            df[column] = df[column].transform(lambda i: typ.names[i])  # 转换为名称
        elif isinstance(typ, Sequence) and isinstance(typ.feature, ClassLabel): # 如果是序列分类标签
            df[column] = df[column].transform(lambda x: [typ.feature.names[i] for i in x]) # 转换为名称列表
    display(HTML(df.to_html())) # 显示DataFrame
````

````python
show_random_elements(dataset["train"], num_examples=5) # 显示训练集中的5个随机样本
````

````python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained(model_name_or_path, trust_remote_code=True) # 加载分词器
````

````python
# tokenize_func 函数
def tokenize_func(example, tokenizer, ignore_label_id=-100):
    """
    对单个数据样本进行tokenize处理。

    参数:
    example (dict): 包含'content'和'summary'键的字典，代表训练数据的一个样本。
    tokenizer (transformers.PreTrainedTokenizer): 用于tokenize文本的tokenizer。
    ignore_label_id (int, optional): 在label中用于填充的忽略ID，默认为-100。

    返回:
    dict: 包含'tokenized_input_ids'和'labels'的字典，用于模型训练。
    """

    # 构建问题文本
    question = prompt_text + example['content'] # 添加全局指令文本和内容
    if example.get('input', None) and example['input'].strip(): # 如果有额外输入，则添加
        question += f'\n{example["input"]}'

    # 构建答案文本
    answer = example['summary'] # 答案文本

    # 对问题和答案文本进行tokenize处理
    q_ids = tokenizer.encode(text=question, add_special_tokens=False) # 不添加特殊标记
    a_ids = tokenizer.encode(text=answer, add_special_tokens=False) # 不添加特殊标记

    # 如果tokenize后的长度超过最大长度限制，则进行截断
    if len(q_ids) > max_input_length - 2:  # 保留空间给gmask和bos标记
        q_ids = q_ids[:max_input_length - 2]
    if len(a_ids) > max_output_length - 1:  # 保留空间给eos标记
        a_ids = a_ids[:max_output_length - 1]

    # 构建模型的输入格式
    input_ids = tokenizer.build_inputs_with_special_tokens(q_ids, a_ids) # 构建输入ID序列
    question_length = len(q_ids) + 2  # 加上gmask和bos标记

    # 构建标签，对于问题部分的输入使用ignore_label_id进行填充
    labels = [ignore_label_id] * question_length + input_ids[question_length:] # 问题部分填充，答案部分保留

    return {'input_ids': input_ids, 'labels': labels} # 返回tokenize后的输入ID和标签
````

````python
column_names = dataset['train'].column_names # 获取数据集的列名
tokenized_dataset = dataset['train'].map( # 对训练集进行tokenize处理
    lambda example: tokenize_func(example, tokenizer),
    batched=False, 
    remove_columns=column_names
)
````

````python
tokenized_dataset = tokenized_dataset.shuffle(seed=seed) # 打乱数据集
tokenized_dataset = tokenized_dataset.flatten_indices() # 展平索引以提高访问速度
print(tokenized_dataset)
````

````python
import torch
from typing import List, Dict, Optional

# DataCollatorForChatGLM 类
class DataCollatorForChatGLM:
    """
    用于处理批量数据的DataCollator，尤其是在使用 ChatGLM 模型时。

    该类负责将多个数据样本（tokenized input）合并为一个批量，并在必要时进行填充(padding)。

    属性:
    pad_token_id (int): 用于填充(padding)的token ID。
    max_length (int): 单个批量数据的最大长度限制。
    ignore_label_id (int): 在标签中用于填充的ID。
    """

    def __init__(self, pad_token_id: int, max_length: int = 2048, ignore_label_id: int = -100):
        """
        初始化DataCollator。

        参数:
        pad_token_id (int): 用于填充(padding)的token ID。
        max_length (int): 单个批量数据的最大长度限制。
        ignore_label_id (int): 在标签中用于填充的ID，默认为-100。
        """
        self.pad_token_id = pad_token_id # 填充token ID
        self.ignore_label_id = ignore_label_id # 忽略标签ID
        self.max_length = max_length # 最大长度限制

    def __call__(self, batch_data: List[Dict[str, List]]) -> Dict[str, torch.Tensor]:
        """
        处理批量数据。

        参数:
        batch_data (List[Dict[str, List]]): 包含多个样本的字典列表。

        返回:
        Dict[str, torch.Tensor]: 包含处理后的批量数据的字典。
        """
        # 计算批量中每个样本的长度
        len_list = [len(d['input_ids']) for d in batch_data] # 计算每个样本的长度
        batch_max_len = max(len_list)  # 找到最长的样本长度

        input_ids, labels = [], [] # 初始化输入和标签列表
        for len_of_d, d in sorted(zip(len_list, batch_data), key=lambda x: -x[0]): # 按长度从大到小排序
            pad_len = batch_max_len - len_of_d  # 计算需要填充的长度
            # 添加填充，并确保数据长度不超过最大长度限制
            ids = d['input_ids'] + [self.pad_token_id] * pad_len # 填充input_ids
            label = d['labels'] + [self.ignore_label_id] * pad_len # 填充labels
            if batch_max_len > self.max_length: # 如果超过最大长度限制，则进行截断
                ids = ids[:self.max_length]
                label = label[:self.max_length]
            input_ids.append(torch.LongTensor(ids)) # 转换为LongTensor并添加到列表
            labels.append(torch.LongTensor(label)) # 转换为LongTensor并添加到列表

        # 将处理后的数据堆叠成一个tensor
        input_ids = torch.stack(input_ids) # 堆叠input_ids
        labels = torch.stack(labels)  # 堆叠labels

        return {'input_ids': input_ids, 'labels': labels} # 返回处理后的批量数据
````

````python
# 准备数据整理器
data_collator = DataCollatorForChatGLM(pad_token_id=tokenizer.pad_token_id)
````

## 加载模型

````python
from transformers import AutoModel, BitsAndBytesConfig

_compute_dtype_map = { # 计算时的数据类型映射
    'fp32': torch.float32,
    'fp16': torch.float16,
    'bf16': torch.bfloat16
}

# QLoRA 量化配置
q_config = BitsAndBytesConfig(load_in_4bit=True,
                              bnb_4bit_quant_type='nf4', # 量化类型
                              bnb_4bit_use_double_quant=True, # 使用双量化
                              bnb_4bit_compute_dtype=_compute_dtype_map['bf16']) # 计算时的数据类型
# 加载量化后模型
model = AutoModel.from_pretrained(model_name_or_path, 
                                  quantization_config=q_config, 
                                  device_map='auto', # 自动分配显存
                                  trust_remote_code=True) 

model.supports_gradient_checkpointing = True  # 支持启用梯度检查点
model.gradient_checkpointing_enable() # 启用梯度检查点
model.enable_input_require_grads() # 启用输入梯度

model.config.use_cache = False  # silence the warnings. Please re-enable for inference!
````

````python
from peft import TaskType, LoraConfig, get_peft_model, prepare_model_for_kbit_training
from peft.utils import TRANSFORMERS_MODELS_TO_LORA_TARGET_MODULES_MAPPING

kbit_model = prepare_model_for_kbit_training(model) # 准备模型以进行k-bit训练
target_modules = TRANSFORMERS_MODELS_TO_LORA_TARGET_MODULES_MAPPING['chatglm'] # 获取ChatGLM的目标模块
print('target_modules:', target_modules)
````

````python
lora_config = LoraConfig( # LoRA配置
    target_modules=target_modules, # 目标模块
    r=lora_rank, # LoRA秩 值是16，现在改为8
    lora_alpha=lora_alpha, # LoRA alpha值是32
    lora_dropout=lora_dropout, # LoRA Dropout率
    bias='none', # 不使用偏置
    inference_mode=False, # 不是推理模式
    task_type=TaskType.CAUSAL_LM # 因果语言模型任务
)
````

````python
qlora_model = get_peft_model(kbit_model, lora_config) # 获取LoRA模型
qlora_model.print_trainable_parameters() # 打印可训练参数
````

### QLoRA 微调模型

````python
train_epochs = 50                           
output_dir = f"models/{model_name_or_path}-epoch{train_epochs}" # 输出目录
````

````python
output_dir
````

````python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments( # 训练参数
    output_dir=output_dir,                            # 输出目录
    per_device_train_batch_size=2,                     # 每个设备的训练批量大小, 原来是8，gpu显存不够，改为2
    gradient_accumulation_steps=1,                     # 梯度累积步数
    learning_rate=1e-3,                                # 学习率
    num_train_epochs=train_epochs,                     # 训练轮数
    lr_scheduler_type="linear",                        # 学习率调度器类型
    warmup_ratio=0.1,                                  # 预热比例, 值原来是0.1
    logging_steps=1,                                 # 日志记录步数
    save_strategy="steps",                             # 模型保存策略
    save_steps=10,                                    # 模型保存步数
    optim="adamw_torch",                               # 优化器类型
    fp16=False,                                        # 是否使用混合精度训练,原来是True,改为False，使用bf16
)
````

````python
trainer = Trainer(
        model=qlora_model, # 训练模型
        args=training_args, # 训练参数
        train_dataset=tokenized_dataset, # 训练数据集
        data_collator=data_collator # 数据整理器
    )
````

````python
trainer.train()
````

````python
trainer.model.save_pretrained(output_dir)
````
